```python
# -----------------------------
# Dataset
# -----------------------------
dataset = [
    {"id": 1, "text": "I love this product", "sentiment": "Positive"},
    {"id": 2, "text": "Very bad experience", "sentiment": "Negative"},
    {"id": 3, "text": "Amazing quality", "sentiment": "Positive"},
    {"id": 4, "text": "Not worth the money", "sentiment": "Negative"},
    {"id": 5, "text": "Excellent performance", "sentiment": "Positive"},
    {"id": 6, "text": "Terrible support", "sentiment": "Negative"},
    {"id": 7, "text": "Happy with purchase", "sentiment": "Positive"},
    {"id": 8, "text": "Waste of time", "sentiment": "Negative"},
    {"id": 9, "text": "Good value", "sentiment": "Positive"},
    {"id": 10, "text": "Disappointed", "sentiment": "Negative"}
]

# -----------------------------
# 1. From Scratch (Simple Naive Bayes)
# -----------------------------
from collections import defaultdict

# Count words in positive and negative reviews
word_counts = {"Positive": defaultdict(int), "Negative": defaultdict(int)}
class_counts = {"Positive": 0, "Negative": 0}

for item in dataset:
    sentiment = item["sentiment"]
    class_counts[sentiment] += 1
    for word in item["text"].lower().split():
        word_counts[sentiment][word] += 1

# Prior probabilities
total = len(dataset)
prior = {c: class_counts[c]/total for c in class_counts}

# Function to predict using simple Naive Bayes
def predict_nb(text):
    words = text.lower().split()
    scores = {}
    for c in ["Positive", "Negative"]:
        score = prior[c]
        for w in words:
            # Laplace smoothing
            word_prob = (word_counts[c].get(w, 0) + 1) / (class_counts[c] + 2)
            score *= word_prob
        scores[c] = score
    return max(scores, key=scores.get)

# From Scratch Predictions
scratch_results = []
for item in dataset:
    pred = predict_nb(item["text"])
    scratch_results.append(pred)

# -----------------------------
# 2. Predefined Model (Hugging Face)
# -----------------------------
from transformers import pipeline

classifier = pipeline("sentiment-analysis")

predefined_results = []
for item in dataset:
    result = classifier(item["text"])[0]
    predefined_results.append(result['label'])

# -----------------------------
# Display Results Side by Side
# -----------------------------
print(f"{'ID':<3} {'Text':<25} {'Actual':<10} {'Scratch':<10} {'Hugging Face':<15}")
print("-"*70)
for i, item in enumerate(dataset):
    print(f"{item['id']:<3} {item['text']:<25} {item['sentiment']:<10} {scratch_results[i]:<10} {predefined_results[i]:<15}")
```
